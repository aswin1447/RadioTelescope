```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal

dat = np.fromfile("200411Apr04.bin", dtype="float32")
```

```python
dat=np.nan_to_num(dat)
```

```python
#dat=dat[1::2] + 1j*dat[0::2]
dat = dat.astype(np.float32).view(np.complex64)
```

```python
plt.specgram(dat, NFFT=1024, Fs=2400000,Fc=100600000)
```

```python
plt.specgram(dat, NFFT=1024, Fs=2400000,Fc=100600000)
plt.title("PSD of 'signal' loaded from file")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.psd(dat, NFFT=1024, Fs=1000000,Fc=100600000)
plt.show()
```

```python
plt.scatter(np.real(dat[0:100000]), np.imag(dat[0:100000]))
plt.title("Constellation of the 'signal' loaded from file")
plt.show()
```

```python
plt.scatter(np.real(dat[0:1000]), np.imag(dat[0:1000]))
plt.title("Constellation of the 'signal' loaded from file")
plt.show()
```
