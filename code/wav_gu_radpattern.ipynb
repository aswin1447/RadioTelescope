```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

a=10.5
b=8
la=21


th=np.linspace(0,89,180)*(np.pi/180)

vx=(a/la)*np.sin(th)
vy=(b/la)*np.sin(th)

k=np.sqrt(1-(la/(2*a))**2)

cE=(1+k*cos(th))/(k+1)

cH=(k+cos(th))/(k+1)

gE=np.square(np.absolute(cE*np.sinc(vy)))

gH= np.absolute(cH*np.square(np.cos(3.14*vx)/(1-4*np.square(vx))))



polar(th,gE)
polar(th,gH)
plt.show()


```
